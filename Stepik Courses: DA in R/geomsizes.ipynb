```r
# Student's criteria
df <- iris
dfr <- subset(df, Species != "setosa")
library(ggplot2)
```

```r
ggplot(dfr, aes(x = Sepal.Length)) +
  geom_histogram(fill = "white", col = "black", binwidth = 0.4) +
  facet_grid(Species ~ .)
options(repr.plot.width = 5, repr.plot.height = 5, repr.plot.res = 1366)
```
